```python
# Creating a new DataFrame df2
df2 = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12]
}, index=['row1', 'row2', 'row3', 'row4'])

# loc: label-based selection
df2.loc['row2']
```

```python
# iloc: integer position-based selection
df2.iloc[1]
```

```python
# Introducing missing values
df2.loc['row2', 'B'] = np.nan

# Drop rows with NaN
print(df2.dropna())

# Fill NaN with a value
print(df2.fillna(0))

# Check where NaN exists
print(df2.isna())
```

```python
# Introduce duplicate rows
df3 = pd.DataFrame({'A': [1, 2, 2, 3], 'B': [4, 5, 5, 6]})

# Check duplicates
print(df3.duplicated())

# Drop duplicate rows
print(df3.drop_duplicates())
```

```python
# Example DataFrame
df4 = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Red']})

# One-hot encoding
encoded_df = pd.get_dummies(df4, columns=['Color'])
print(encoded_df)
```

```python
# Select rows where A > 0
print(df2[df2['A'] > 0])
```

```python
# Concatenate two DataFrames
df5 = pd.DataFrame({'A': [1, 2]})
df6 = pd.DataFrame({'A': [3, 4]})
concat_df = pd.concat([df5, df6])
print(concat_df)
```
